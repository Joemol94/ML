```python
import pandas as pd

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Read the iris dataset
iris=pd.read_csv("/home/user/Desktop/notebook/iris.csv",index_col='Item')

#size of the data frame
print(iris.shape)
```

```python
#View the first 5 rows of the iris dataset
iris.head()
```

```python
#Bar plot

sns.barplot(x=iris['Species'], y=iris['Sepal.Length'])
```

```python
#Removing the species column
da=iris[iris.columns[:-1]]
da.head()
```

```python
#PLotting a rough line plot to visualize all the 4 attributes irrespective of species
plt.figure(figsize=(14,6))
plt.title("Line chart for iris data")

sns.lineplot(data=da)
```

```python
#Visualizing a heat map
plt.figure(figsize=(14,7))
plt.title("Heatmaps")
sns.heatmap(data=da[0:10],annot=True)
```

```python
#Scatter plot to visualize the correlation between 2 variables
#Here the 2 variables are petal length and sepal length 
sns.scatterplot(x=iris['Petal.Length'], y=iris['Sepal.Length'], hue=iris['Species'])
```

We can see from the above that when petal length increases sepal length also increases in versicolor and virginica species.
In otherwords it shows a positive correlation.

```python
#create a pairplot to plot across different attributes. 
sns.pairplot(iris, hue='Species')
```

```python
#visualizing a violin plot
sns.violinplot(x=iris['Species'],y=iris['Sepal.Length'],data=iris)
```

```python
iris_set=iris[iris.Species=='setosa']
```

```python
iris_ver=iris[iris.Species=='versicolor']
```

```python
iris_vir=iris[iris.Species=='virginica']
```

```python
plt.title("Histograms for Sepal length by species")
sns.distplot(iris_set['Sepal.Length'],label="setosa",kde=False)
sns.distplot(iris_ver['Sepal.Length'],label="versicolor",kde=False)
sns.distplot(iris_vir['Sepal.Length'],label="virginica",kde=False)
plt.legend()
```

```python
plt.title("KDE plot for Sepal length by species")
sns.kdeplot(iris_set['Sepal.Length'],label="Iris setosa",shade=True)
sns.kdeplot(iris_ver['Sepal.Length'],label="Iris versicolor",shade=True)
sns.kdeplot(iris_vir['Sepal.Length'],label="Iris virginica",shade=True)
plt.legend()
```
